```python
import pandas as pd

donne = pd.read_csv('pratices/vehicules.csv')
data = donne.copy()

liste = ['image_url','description','posting_date','size', 'state', 'VIN', 'region_url', 'region', 'url', 'long',
       'id', 'lat', 'Unnamed: 0', 'model']

data = data[data.price > 1500]
data.drop(liste, axis=1, inplace=True)
data = data.sample(150000, random_state=42)
data
```

### First model to test the base preprocessing

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import learning_curve, train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor, AdaBoostRegressor
import numpy as np
from sklearn.svm import SVR
from sklearn.linear_model import SGDRegressor
import matplotlib.pyplot as plt
from sklearn.pipeline import make_pipeline
from sklearn.feature_selection import SelectKBest
from sklearn.preprocessing import StandardScaler,LabelEncoder,PolynomialFeatures
from sklearn.impute import SimpleImputer, KNNImputer
```

## Modelisation 

```python
def encodage(data):
    for strings in data.select_dtypes('object'):
        data[strings] = (data[strings]).astype("category").cat.codes
encodage(data)


for dt in data.columns:
    data[dt] = data[dt].fillna(data[dt].mean())
data
```

```python
import seaborn as sb
sb.pairplot(data, hue='condition', data=data)
```

```python
train, test = train_test_split(data, random_state=42, test_size=0.65)

X_train, y_train = train.drop('price', axis=1), train['price']
X_test, y_test = test.drop('price', axis=1), test['price']
```

### Modelisation 

```python
models = [LinearRegression() , SVR() , AdaBoostRegressor(), ]
model= SGDRegressor()
model.fit(X_train, y_train)
model.score(X_train, y_train)
```

```python
def modelisation(model=model, X=X_train, y=y_train, x=X_test, Y=y_test):
    model.fit(X, y)
    
    print(model.score(x, Y))
    
    # Learning curve now
    
    train_sizes, test_score, val_score = learning_curve(model, X, y, train_sizes=np.linspace(.1, 1.0, 10), random_state=42)
    
    plt.figure(figsize=(16,8))
    plt.plot(train_sizes, test_score.mean(axis=1), label="Entrainement")
    plt.plot(train_sizes, val_score.mean(axis=1), label="Validation")
    plt.legend()
    
modelisation()
```
